```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
def plot_patient_preds(p_csv_filename, p_name):
    predictions_file = pd.read_excel('../../../../cidia_scripts/' + p_csv_filename)
    y_axis = predictions_file["predictions"]
    x_axis = np.linspace(0,100,100)
    plt.plot(x_axis, y_axis, 'bo', label='', scaley = False)
    plt.title('Grau de Confianca por fatia do paciente ' + p_name)
    plt.show()
```

```python
plot_patient_preds('preds_p007.xlsx', 'NEG-007')
```
